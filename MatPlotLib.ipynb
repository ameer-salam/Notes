## Line Plot


````python
import sympy as sp
from sympy import symbols, Eq
import numpy as np
import matplotlib
from matplotlib import pyplot as plt

x=np.arange(1,22)
print(x)

y=symbols("y")
y=2*x
print(y)

plt.title("X and Y relation")
plt.xlabel("X values ->")
plt.ylabel("Y values ->")
#line patterns supported are
#'-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'
plt.plot(x, y, color="green", linestyle=":", linewidth=3)
plt.plot(y, x*2, color="red", linestyle="dashdot", linewidth=4)
#plt.grid(True)
plt.show()

w=symbols("w")
z=symbols("z")

w=x**2
z=x**3

plt.subplot(1,2,1)
plt.plot(x, w, color='g', linestyle='dashdot', linewidth=2)

plt.subplot(1,2,2)
plt.plot(x, w, color='b', linestyle=':', linewidth=3)
plt.show()
````

## Bar Plot
* requires creation of library


````python
#plotting a Bar graph using Dictionary
students={'Ameer':98, "Gagan Raj RP":32, "Charan H M":45, "Abhishek VS":23}
name=list(students.keys())
values=list(students.values())
plt.bar(name, values)
plt.title("Students Result")
plt.xlabel("Marks")
plt.ylabel("Student Names")
plt.show()

# to make the bar horizonata and red in colour
plt.barh(name, values, color="r")
plt.title("Students Result")
plt.xlabel("Student Names")
plt.ylabel("Marks")
plt.show()
````

## Scatter Plot
it is used to understand the distributionn between two numerical entiries which are represented in the form of data points.

````python
import random
x = random.sample(range(1, 46), 10)
y = random.sample(range(34, 345), 10)
a = list(range(1, 11))
plt.scatter(x,a, color="g", marker="*", s=150)
plt.scatter(y,a, color="r", marker=".", s=150)
plt.title("Scatter Plot graph")
plt.xlabel("X - Values ->")
plt.ylabel("A - Values ->")

plt.show()

````

## Box Plot

## Violin Plot, Pie Chart and Donut Chart
* Violin plot is as same as the box plot but with curved edges to represnt the mean value position
* pi chart helps us understant the frequency or percentage of different categorical values

````python
#Violin plot
one=[2,3,4,5,6,8,5,6,7,8]
two=[1,3,5,2,4,3,1,7,3,1]
three=[5,7,3,6,9,1,2,4,1,4]

data=list([one, two, three])
plt.violinplot(data, showmedians=True) #also shows the median
plt.show()
````

````python
#pie-chart
fruit=['Apple', 'Mango', 'Banana', 'Sapota', 'Papaaya'] #categorical entity
values=[20, 100, 70, 90, 60] #numerical entity

plt.pie(values, labels=fruit, autopct='%0.5f%%', colors=['red', 'Gold', 'yellow', 'Brown', 'orange'])
plt.show()
````

````python
#donut-chart
fruit=['Apple', 'Mango', 'Banana', 'Sapota', 'Papaaya'] #categorical entity
values=[20, 100, 70, 90, 60] #numerical entity

plt.pie(values, labels=fruit, autopct='%0.5f%%', colors=['red', 'Gold', 'yellow', 'Brown', 'orange'], radius=2)
plt.pie([1], colors=['w'], radius=1)
plt.show()
````
